<H3>Question 1: Find the ten countries with the most projects.</h3>

```python
# A quick view of the top ten countries
df.countryshortname.value_counts().head(10)
```

```python
#Plot as a bar graph for fun.
def top_ten_countries(location = 'data/world_bank_projects.json'):
    '''Function parses desired json file and returns a pandas dataframe of the top 10 countries with the 
    most projects. Plots a bar graph.'''
    
    #INPUT - JSON DATA LOCATION
    #OUPTUT - DATAFRAME AND BAR GRAPH OF TOP 10 COUNTRIES WITH THE MOST PROJECTS AND 
    
    #parse JSON data into a pandas DataFrame
    df1 = pd.read_json(location)
    
    # .loc[] - slice dataframe using all rows and only columns 'countryname' and 'countryshortname
    # .groupby().count() - group dataframe by countryshortname and count how many instances of each country there are
    # .sort_values - sort data in descending order
    
    df1 = df1.loc[:,['countryname','countryshortname']].groupby('countryshortname').count().sort_values('countryname',ascending = False)
    
    #return dataframe and bar graph of the top 10 countries with the most projects
    return df1[0:10], df1[0:10].plot(kind = 'bar',rot = 45, figsize=(13,8), legend = False)

#call function above and answer question
df1, q1 = top_ten_countries()


################# BAR GRAPH ##########################################################
#Change 'Yemen, Repbulic of' to just 'Yemen'
labels = [label.get_text() for label in q1.get_xticklabels()]
labels[4] = 'Yemen'

#Label the graph
q1.set_xticklabels(labels)
q1.set_title('Top 10 Countries with the Most Projects')
q1.set_xlabel('Country')
q1.set_ylabel('Number of Projects')

#Remove grid and Y-axis labels
q1.grid(False)
q1.get_yaxis().set_ticks([])

#Label each bar with its corresponding value
rects = q1.patches
labels = list(df1.countryname)

for rect, label in zip(rects,labels):
    height = rect.get_height()
    q1.text(x = rect.get_x() + rect.get_width()/2, 
            y = height - 1 , 
            s = label,
            ha = 'center', 
            va = 'bottom',
            color = 'white')

########################################################################################

plt.show()

df1.rename(columns={'countryname': 'count'}, inplace=True)
df1
```

<h3>Question 2: Find the top 10 major project themes (using column 'mjtheme_namecode')</h3>

```python
def top_ten_projects(location = 'data/world_bank_projects.json'):
    '''This function reads JSON data into a datafram (df) and runs the "count" function to output the top 
    10 project themes in the df'''
    
    #INPUT -> JSON DATA LOCATION
    #OUTPUT -> DATAFRAME AND BAR GRAPH WITH THE TOP 10 PROJECT THEMES
    
    #Open the file and alias it as object fh
    with open(location) as js:
        
        #parse the JSON file into a Python object
        df2 = json.load(js)
       
        #parse the data into a pandas df
        df2 = json_normalize(df2,"mjtheme_namecode")
        print(df2.columns)
        
        # .groupby().count() -> group df2 by project names and count how many instances there are of each
        # .sort_values() -> sort the count values in descending order
        df2 = df2.groupby('name').count().sort_values('code',ascending = False)
        #rename "code" column name to 'Count'.  Can't rename 'name' and I'm not sure why
        df2.rename(columns={'code': 'Count'}, inplace=True)
    #return the datframe 

    return df2[0:10]


#call the above function and answer the question

top_ten_projects()
```

<h3>Question 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.</h3>

```python
#Create the dataframe again...this time with the missing project names filled in
# load json as string
data = json.load((open('data/world_bank_projects.json')))

# use normalization to create tables from nested element
df3 = json_normalize(data,'mjtheme_namecode')

# Some entries have only the code and the name is missing.
# Create a dataframe with the missing names filled in
df3 = df3.replace('',np.nan).groupby('code').ffill().bfill()

# Find the top 10 major project themes
df3.name.value_counts()[:10]
```
